```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diamonds.csv', sep = ',')
df.head(10)
```

```python
df.info()
```

```python
#изменение размера занимаемой памяти
df['carat'] = df['carat'].astype('float32')
df['depth'] = df['depth'].astype('float32')
df['table'] = df['table'].astype('float32')
df['x'] = df['x'].astype('float32')
df['y'] = df['y'].astype('float32')
df['z'] = df['z'].astype('float32')

df['Unnamed: 0'] = df['Unnamed: 0'].astype('int32')
df['price'] = df['price'].astype('int32')

df.info()
```

```python
df.describe()
```

```python
df.describe(include=['object'])
```

```python
df.sample(300).plot.scatter(x='price', y='carat',figsize=(14,5))
df.plot.scatter(x='price', y='carat',figsize=(14,5))
```
