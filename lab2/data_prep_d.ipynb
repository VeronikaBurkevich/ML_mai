```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('diamonds.csv')
df.color.unique(), df.cut.unique(), df.clarity.unique()
```

```python
d1 = {'E' : 0, 'I' : 1, 'J' : 2, 'H' : 3, 'F' : 4, 'G' : 5, 'D' : 6}
df['color'] = df['color'].map(d1)
d2 = {'Ideal' : 0, 'Premium' : 1, 'Good' : 2, 'Very Good' : 3, 'Fair' : 4}
df['cut'] = df['cut'].map(d2)
d3 = {'SI2':0, 'SI1':1, 'VS1':2, 'VS2':3, 'VVS2':4, 'VVS1':5, 'I1':6, 'IF':7}
df['clarity'] = df['clarity'].map(d3)

df.color.unique(), df.cut.unique(), df.clarity.unique()
```

```python
#Тепловая карта
import matplotlib.pyplot as plt
import seaborn as sns
fig = plt.figure(figsize= (14,5))
sns.heatmap(df.dropna().corr(), cmap = 'Blues')
```

```python
df.drop(['Unnamed: 0', 'depth', 'clarity', 'color'], axis='columns', inplace=True)
df.info()
```

```python
#изменение размера занимаемой памяти
df['carat'] = df['carat'].astype('float32')
df['table'] = df['table'].astype('float32')
df['x'] = df['x'].astype('float32')
df['y'] = df['y'].astype('float32')
df['z'] = df['z'].astype('float32')

df['cut'] = df['cut'].astype('int32')
df['price'] = df['price'].astype('int32')

df.info()
```
